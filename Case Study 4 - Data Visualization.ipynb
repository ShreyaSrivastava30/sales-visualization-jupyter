```python
## 1. Compare Sales by region for 2016 with 2015 using bar chart.
```

```python
sales[['Region','Sales2015','Sales2016']].groupby('Region').sum().plot(kind='bar')
plt.ylabel('Sales')
plt.title('Sales by Region in 2015 & 2016')
plt.show()
```

```python
## 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.
```

```python
region_2016_sales = sales[['Region','Sales2016']].groupby('Region').sum()
```

```python
region_2016_sales
```

```python
plt.pie(region_2016_sales.Sales2016,autopct="%1.0f%%",labels=["Central","East","West"],shadow=True,colors=['r', 'g', 'b'])
plt.title("Contributing factors to the sales for each region in 2016")
plt.show()
```

```python
## 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers.
```

```python
sales[['Tier','Region','Sales2015','Sales2016']].groupby(['Region','Tier']).sum().plot(kind='bar',figsize=(15,8))
plt.ylabel("Sales")
plt.title("Total sales of 2015 and 2016 with respect to Region and Tiers")
plt.show()
```

```python
## 4. In East region, which state registered a decline in 2016 as compared to 2015?
```

```python
sales[sales.Region=='East'][['State','Sales2015','Sales2016']].groupby('State').sum().plot(kind='bar',figsize=(15,8))
plt.ylabel("Sales")
plt.title("Total sales of 2015 and 2016 in East Region with respect to State")
plt.show()
```

```python
## 5. In all the High tier, which Division saw a decline in number of units sold in 2016 compared to 2015?
```

```python
sales[sales.Tier=='High'][['Division','Units2015','Units2016']].groupby('Division').sum().plot(kind='bar',figsize=(18,8))
plt.ylabel("Units Sold")
plt.title("Total number of units sold in 2015 and 2016 for High Tier with respect to Division")
plt.show()
```

```python
## 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. 
##    The Quarters are based on months and defined as -
##          â€¢ Jan - Mar : Q1
##          â€¢ Apr - Jun : Q2
##          â€¢ Jul - Sep : Q3
##          â€¢ Oct - Dec : Q4
```

```python
quarter = []
for x in sales.Month :
    if x in ["Jan","Feb","Mar"]:
        quarter.append("Q1")
    elif x in ["Apr","May","Jun"]:
        quarter.append("Q2")
    elif x in ["Jul","Aug","Sep"]:
        quarter.append("Q3")
    else:
        quarter.append("Q4")
```

```python
sales['Qtr'] = quarter
```

```python
sales
```

```python
## 7. Compare Qtr wise sales in 2015 and 2016 in a bar plot.
```

```python
sales[['Qtr','Sales2015','Sales2016']].groupby('Qtr').sum().plot(kind='bar',figsize=(15,5))
plt.ylabel("Sales")
plt.title("Quarter wise sales in 2015 and 2016")
plt.show()
```

```python
## 8. Determine the composition of Qtr wise sales in 2016 with regards to all the Tiers in a pie chart.
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(sales[['Sales2015', 'Sales2016', 'Units2015', 'Units2016']].corr(), 
            annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Sales and Units (2015 vs 2016)")
plt.show()
```

# Observations from Correlation Heatmap:

- Sales2015 and Units2015 have a high correlation (0.92) â€” as expected, more units sold results in more revenue.
- Sales2016 and Units2016 also show a strong positive correlation (0.95).
- Year-over-year correlations are moderate (~0.6), indicating some consistency but also potential shifts in sales dynamics between 2015 and 2016.
- This correlation analysis confirms that unit sales significantly drive total revenue.

```python
# Select numeric columns for pairplot
numeric_cols = ['Sales2015', 'Sales2016', 'Units2015', 'Units2016']

# Create pairplot
sns.pairplot(sales[numeric_cols], height=2.5)
plt.suptitle("Pairplot of Sales and Units (2015 vs 2016)", y=1.02)
plt.show()
```

Distributions
